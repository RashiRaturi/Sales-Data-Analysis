```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## LOADING DATASET

```python
df = pd.read_csv('sales_data.csv')
```

```python
df.head(5)
```

## LINE PLOT 
### Revenue vs ID

```python
x = df['ID']
y = df['Revenue']
```

```python
plt.plot(x,y)

plt.title('Sales Data')
plt.xlabel('ID')
plt.ylabel('Revenue')
plt.xticks(np.arange(0, 101, 10))
plt.yticks(np.arange(50, 201, 10))
plt.grid(True)

plt.show()
```

## SCATTER PLOT (QUANTITITY VS REVENUE)
### Each point represents a transaction.

```python
x = df['Quantity']
y = df['Revenue']
```

```python
df['Product'].unique()       # checking how many unique products are there
```

```python
colors = {'B': 'blue', 'A': 'green', 'C': 'red'}  # creating a dictionary for color coding

plt.title('Revenue by Quantity')
plt.xlabel('Quantity')
plt.ylabel('Revenue')

# Creating scatter point with color coded points based on product
for prod_type, color in colors.items():
    subset_df = df[df['Product'] == prod_type]
    plt.scatter(subset_df['Quantity'], subset_df['Revenue'], c=color, label=f'Product {prod_type}')

plt.legend()

```

## BAR CHART (REVENUE VS PRODUCT)
### Revenue for each category of product

```python
total_revenue = df.groupby('Product')['Revenue'].sum()  # grouping by product and summing up the revenue
total_revenue
```

```python
x = total_revenue.index
y = total_revenue.values
```

```python
plt.bar(x,y)

plt.title('Total Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Revenue')

plt.show()
```
